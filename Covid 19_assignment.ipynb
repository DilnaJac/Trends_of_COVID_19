# Forecasting for confirmed cases

```python
confirmed.columns = ['ds','y']
```

```python
confirmed
```

```python
from prophet import Prophet
```

```python
m = Prophet()
```

```python
m.fit(confirmed)
```

```python
future = m.make_future_dataframe(periods=7)
```

```python
future
```

```python
forecast = m.predict(future)
```

```python
forecast
```

```python
forecast[['ds','yhat','yhat_upper','yhat_lower']]
```

```python
m.plot(forecast)
plt.show()
```
